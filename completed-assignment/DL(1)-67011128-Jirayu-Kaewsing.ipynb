```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
```

Random data generation

```python
# Generate random data

x_data = np.random.randn(1000, 3) # 1000 samples with 3 features
w_real = [0.3, 0.5, 0.1] # weights
b_real = -0.2 # bias
y_data = np.matmul(w_real, x_data.T) + b_real # output
```

Split the data into training and test sets

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.1, random_state = 42)
```

Model definition

```python
# Define a model
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=[3]),
    tf.keras.layers.Dense(units=2),
    tf.keras.layers.Dense(units=1)
])
```

Model compilation

```python
LEARNING_RATE = 1e-2

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=LEARNING_RATE), 
    loss='mse'
)
```

Model training

```python
BATCH_SIZE = 100
N_ITERATIONS = 100

model.fit(
    x_train, 
    y_train, 
    batch_size=BATCH_SIZE, 
    epochs=N_ITERATIONS
)
```

Model architecture summary

```python
model.summary()
```

Model's internal parameters (weight, bias)

```python
weights = model.get_weights()
print(f'Weight: {weights[0]}, Bias: {weights[1]}')
```

```python
# Evaluate loss (test_loss) on the test set
test_loss = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Loss (MSE): {test_loss:.6f}\n")
```

```python
# Make predictions (evaluate y_pred) on x_test
y_pred = model.predict(x_test, verbose=0)
```

```python
# Print sample number, x_test, y_test, and y_pred
print(f"{'Sample':<8} | {'x_test (features)':<30} | {'y_test (actual)':<15} | {'y_pred (predicted)':<15}")
print("-" * 80)

for i in range(10):
    
    features = np.round(x_test[i], 4)
    actual = round(y_test[i], 4)
    
    predicted = round(y_pred[i][0], 4) 
    
    print(f"{i+1:<8} | {str(features):<30} | {actual:<15} | {predicted:<15}")
```

```python
test_acc = 1 - test_loss / np.var(y_test)
print(f'Test accuracy: {test_acc}')
```
