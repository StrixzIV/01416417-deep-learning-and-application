```python
import numpy as np
import tensorflow as tf

from IPython.display import Image
```

```python
xy = np.loadtxt('../data/data-01-test-score.csv', delimiter = ',', dtype = np.float32)
xy
```

```python
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape, x_data)
print(y_data.shape, y_data)
```

```python
x_test = np.array([[100, 70, 101]], dtype = np.float32)
x_test
```

Defining model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=(3,))
])

model
```

Compiling model

```python
LEARNING_RATE = 1e-2

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='mean_squared_error'
)
```

Training model

```python
N_ITERATIONS = 2000

history = model.fit(x_data, y_data, epochs=N_ITERATIONS)
print("Training complete! Final Loss:", history.history['loss'][-1])
```

Model architecture summary

```python
model.summary()
```

Model's internal parameters (weight, bias)

```python
weights = model.get_weights()
print(f'Weight: {weights[0]}, Bias: {weights[1]}')
```

Model's prediction on test data

```python
y_pred = model.predict(x_test)
print(f'Predicted score: {y_pred[0][0]}')
```
