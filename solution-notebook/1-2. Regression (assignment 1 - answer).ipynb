```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Generate random data
x_data = np.random.randn(1000, 3) # 1000 samples with 3 features
w_real = [0.3, 0.5, 0.1] # weight
b_real = -0.2 # bias
y_data = np.matmul(w_real, x_data.T) + b_real # output
```

```python
# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.1, random_state = 42)
```

```python
# Define a model
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=[3]),
    tf.keras.layers.Dense(units=2, activation='relu'),
    tf.keras.layers.Dense(units=1)
])
```

```python
# Compile the model
model.compile(optimizer = tf.keras.optimizers.SGD(learning_rate = 0.5), loss = 'mean_squared_error')
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs = 100, batch_size = 100, verbose = 1)
```

```python
# Print the model's summary
model.summary()
```

```python
# Print the weight and bias of the model
# The weights are represented by a 3x2 matrix 
weights = model.get_weights()
print(f'Weight: {weights[0]}, Bias: {weights[1]}')
```

```python
# Evaluate test loss on the test set
test_loss = model.evaluate(x_test, y_test)
print(f'Test loss: {test_loss}')
```

```python
# Make predictions on x_test
y_pred = model.predict(x_test)
```

```python
# Print sample number, x_test, y_test, and y_pred
for i in range(len(x_test)):
    print(f'x_test[{i}]: {x_test[i]}, y_test[{i}]: {y_test[i]}, y_pred[{i}]: {y_pred[i][0]}')
```

```python
# Compute test accuracy
# The formula used here suggests that the test accuracy is calculated 
# by taking 1 and subtracting the ratio of the test loss to the variance of the true labels.
test_acc = 1 - test_loss/np.var(y_test)
print(f'Test accuracy: {test_acc}')
```
